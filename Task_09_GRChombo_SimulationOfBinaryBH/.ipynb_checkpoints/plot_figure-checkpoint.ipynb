```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
!mkdir ./signal_pattren
```

```python
! head data/Weyl4_mode_20.dat
```

```python
data = np.loadtxt('data/Weyl4_mode_20.dat', skiprows=2)
time=data[:,0]
re32,im32 = data[:,1],data[:,2]
re50,im50 = data[:,3],data[:,4]
matplotlib.rc('figure', figsize = (12,8))
matplotlib.rc('font', size = 12)
matplotlib.rc('axes', grid = True)
matplotlib.rc('axes', facecolor = 'white')

_,ax = plt.subplots(2, 1, figsize=(12, 10), sharex=True)
ax[0].plot(time, re32, label='r=32 (Re)', color='blue', alpha=0.8)
ax[0].plot(time, re50, label='r=50 (Re)', color='red', alpha=0.8)
ax[0].set_title('Gravitational Wave Mode (l=2, m=0) - Real Part')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Integral Value')
ax[0].legend()

ax[1].plot(time, np.abs(im32), label='r=32 (|Im|)', color='blue', linestyle='--', alpha=0.7)
ax[1].plot(time, np.abs(im50), label='r=50 (|Im|)', color='red', linestyle='--', alpha=0.7)
ax[1].set_yscale('log')
ax[1].set_title('Absolute Value of Imaginary Part (Log Scale)')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('|Integral Value|')
ax[1].legend()
signature_text = "plot by Wang Ruizhe\nCAS Selection 2026"
plt.gca().text(
    0.98, 0.02, signature_text, 
        transform=plt.gca().transAxes, 
        fontsize=10, 
        color='gray', 
        alpha=0.6,    
        ha='right', 
        va='bottom', 
        fontstyle='italic',
        bbox=dict(facecolor='white', alpha=0.3, edgecolor='none')
)


plt.tight_layout()
plt.subplots_adjust(bottom=0.1) 

plt.savefig('signal_pattern/Weyl4_mode_20_combined.png', dpi=300, bbox_inches='tight')

```
